```python
import matplotlib.pyplot as plt

plt.pie(
    df['gender'].value_counts(),
    labels=df['gender'].value_counts().index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Gender Distribution")
plt.show()
```
