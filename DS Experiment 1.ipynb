**1. Write a program to compute factorial of a given number**

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = int(input("Enter a number: "))
print(f"The factorial of {number} is {factorial(number)}")
```


2. Write a program to display table for a given number


```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_table = int(input("Enter a number to display its multiplication table: "))
display_multiplication_table(num_to_table)

```


3. Write a program to identify the given number is odd or even


```python
def check_odd_even(number):

    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num = int(input("Enter a number: "))
print(check_odd_even(num))

```


4. WAP to identify the entered number is palindrome or not.


```python
def is_palindrome(number):

    # Convert the number to a string to easily reverse it
    num_str = str(number)

    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num = int(input("Enter a number: "))
print(is_palindrome(num))

```


5. Develop an Expert System that provides simple decision-making.

```python
def career_expert_system():
    print("Student Career Path Expert System")

    math = input("Do you like Math? (yes/no): ").lower()
    science = input("Do you like Science? (yes/no): ").lower()
    computers = input("Do you like Computers? (yes/no): ").lower()
    art = input("Do you like Art or Creativity? (yes/no): ").lower()

    if math == "yes" and computers == "yes":
        print("Suggested Career: Software Engineer / Data Scientist")
    elif science == "yes" and math == "yes":
        print("Suggested Career: Doctor / Engineer / Researcher")
    elif art == "yes":
        print("Suggested Career: Designer / Artist / Media Professional")
    elif computers == "yes":
        print("Suggested Career: IT Support / Programmer")
    else:
        print("Suggested Career: Explore your interests and talk to a career counselor")

# Run the expert system
career_expert_system()
```
