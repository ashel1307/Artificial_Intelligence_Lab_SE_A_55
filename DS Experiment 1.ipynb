**1. Write a program to compute factorial of a given number**

```python
def factorial(n):
    """Computes the factorial of a non-negative integer n."""
    if not isinstance(n, int) or n < 0:
        raise ValueError("Factorial is defined only for non-negative integers.")
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
try:
    fact = factorial(number)
    print(f"The factorial of {number} is {fact}")
except ValueError as e:
    print(f"Error: {e}")

```


2. Write a program to display table for a given number


```python
def display_multiplication_table(number):
    """Displays the multiplication table for a given number from 1 to 10."""
    if not isinstance(number, (int, float)):
        raise ValueError("Input must be a number.")

    print(f"\nMultiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
try:
    display_multiplication_table(num)
except ValueError as e:
    print(f"Error: {e}")

```


3. Write a program to identify the given number is odd or even


```python
def check_odd_even(number):
    """Checks if a given number is odd or even."""
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 4
print(check_odd_even(num1))

```


4. WAP to identify the entered number is palindrome or not.


```python
def is_palindrome(number):
    """Checks if a given number is a palindrome."""
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    # Convert the number to a string to easily reverse it
    num_str = str(number)

    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

```
