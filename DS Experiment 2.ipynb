```python
BREADTH FIRST SEARCH
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': ['H'],
    'G': [],
    'H': []
}

def bfs(start):
    visited = set([start])
    queue = [start]
    while queue:
        node = queue.pop(0)
        print(node, end=' ')
        for n in graph[node]:
            if n not in visited:
                visited.add(n)
                queue.append(n)

bfs('A')
print()
```

```python
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G'],
    'D': [],
    'E': ['H'],
    'F': [],
    'G': [],
    'H': []
}

def dfs(start, visited=set()):
    print(start, end=' ')
    visited.add(start)
    for n in graph[start]:
        if n not in visited:
            dfs(n, visited)

print()
dfs('A')
```

### A* Search Algorithm

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

h = {
    'A': 3,
    'B': 2,
    'C': 2,
    'D': 1,
    'E': 1,
    'F': 0
}

def a_star(start, goal):
    open_list = [start]
    g = {start: 0}
    parent = {}

    while open_list:
        current = min(open_list, key=lambda x: g[x] + h[x])

        if current == goal:
            path = [goal]
            while current in parent:
                current = parent[current]
                path.append(current)
            return path[::-1]

        open_list.remove(current)

        for n in graph[current]:
            temp_g = g[current] + 1
            if n not in g or temp_g < g[n]:
                g[n] = temp_g
                parent[n] = current
                open_list.append(n)

print(a_star('A', 'F'))
```

### Minimax Algorithm

```python
def minimax(depth, node, maximizing):
    if depth == 0:
        return node

    if maximizing:
        best = -1000
        for child in node:
            best = max(best, minimax(depth - 1, child, False))
        return best
    else:
        best = 1000
        for child in node:
            best = min(best, minimax(depth - 1, child, True))
        return best

# Example game tree (leaf values)
game_tree = [
    [3, 5, 2],
    [9, 12, 5],
    [4, 6, 8]
]

print(minimax(2, game_tree, True))
```
