```python
import numpy as np
matrix = np.array ([[1,2,1],[3,7,1],[2,2,7]])
```

```python
determinant = np.linalg.det(matrix)

print(f"Determinant: {determinant : .2f} ")
```

```python
matrix_inv = np.linalg.inv(matrix)

print("Inverse Matrix:")
print(matrix_inv)
```

```python
B = np.array([9, -5, -9])

X = np.linalg.solve(matrix, B)

print(f"Solution: x = {X[0] :.0f}, y = {X[1] :.0f}, z = {X[2] :.0f}")
```
